```python
import math
import numpy as np
import pandas as pd

import seaborn as sns
from IPython.display import display
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10,6]

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 15+ Regression Models
# Machine Learning Algorithms
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import ElasticNet,LinearRegression
from sklearn.linear_model import SGDRegressor, ARDRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor,BaggingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.tree import ExtraTreeRegressor, DecisionTreeRegressor
from sklearn.svm import SVR 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
import lightgbm as lgb
# Evaluation Matrix for Regression 
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
#mean_absolute_error(MAE)
#mean_squared_error(MSE)
#r2_score(R-Squared)

import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("/kaggle/input/crop-yield-prediction/crop yield data sheet.xlsx")
data.head()
```

## Missing Vaue Handling

```python
data.columns
```

```python
data.rename(columns={"Rain Fall (mm)":"Rain_Fall_mm",'Nitrogen (N)':'Nitrogen', 'Phosphorus (P)':'Phosphorus','Potassium (K)':'Potassium','Yeild (Q/acre)':'Yeild' },inplace=True)
data.head()
```

```python
data.drop_duplicates()
```

```python
data.Yeild.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
df = data[data.Yeild.isnull() == False]
df.isnull().sum()
```

```python
df.Temperatue.unique()
```

```python
df['Temperatue'] = df['Temperatue'].astype(int)
```

```python
df['Temperatue'].dtype
```

```python
df.info()
```

```python
X = df.drop('Yeild', axis=1)
y = df['Yeild']
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
LR = LinearRegression()
LR.fit(X_train, y_train)

pred1 = LR.predict(X_test)

print("mean_squared_error",round(np.sqrt(mean_squared_error(y_test, pred1)),2))
print("mean_absolute_error",round(np.sqrt(mean_absolute_error(y_test, pred1)),2))
print("r2_score",round(np.sqrt(r2_score(y_test, pred1)),2))
```
